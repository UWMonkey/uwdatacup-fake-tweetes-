```python
import json
import pandas as pd
import numpy as np
```

```python
# read file
with open('train.json', 'r') as myfile:
    data=myfile.read()

    # parse file
obj = json.loads(data)
df = pd.DataFrame(obj)
```

```python
df.head()
```

```python
#replace missing with others first
df["claimant"] = np.where(df["claimant"]=="","other",df["claimant"])
df.head()
```

```python
unique = df["claimant"].value_counts()
unique.head()
```

```python
unique_df=pd.DataFrame(unique)
unique_df.head()
```

```python
unique_df["cum_%"]=unique_df["claimant"]/unique_df["claimant"].sum()*100
unique_df.head()
```

```python
value = unique_df["claimant"]
idx = unique.index
y_1 = pd.DataFrame(np.transpose([idx,value]),columns=["claimant","value"])
```

```python
new_df = pd.merge(df,y_1,on = "claimant")
new_df.head()
```

```python
#create a column for count <=50 then other, else the same
df["new_claimant"] = np.where(new_df["value"]<=30,"other",new_df["claimant"])
df.head()
```

```python
new_df["new_claimant"].value_counts()
```
